```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("UCI_Credit_Card.csv")
df
```

```python
for i in range(25):
    print("Null values in Column",i,":")
    nullCheck=df.iloc[:,i].isnull().sum()
    print(nullCheck)
```

```python
X=df.drop(["ID","default.payment.next.month"],axis='columns')
Y=df["default.payment.next.month"]
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X.iloc[:,12:24]=scaler.fit_transform(X.iloc[:,12:24])
X
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=5)
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn1=KNeighborsClassifier(n_neighbors=3)
knn1.fit(X_train,Y_train)
Y_predicted=knn1.predict(X_test)
print("First Five Test Values:")
print(np.array(Y_test[0:5]))
print("First Five Predicted Values:")
print(Y_predicted[0:5])
print(knn1.score(X_test,Y_test))
```

```python
knn2=KNeighborsClassifier(n_neighbors=5)
knn2.fit(X_train,Y_train)
Y_predicted=knn2.predict(X_test)
print("First Five Test Values:")
print(np.array(Y_test[0:5]))
print("First Five Predicted Values:")
print(Y_predicted[0:5])
print(knn2.score(X_test,Y_test))
```

```python
matrix=metrics.confusion_matrix(Y_predicted,Y_test)
sns.heatmap(matrix, annot=True, linewidths=0.01,cmap="Blues",linecolor="gray", fmt= '.1f')
plt.xlabel("Predicted Values")
plt.ylabel("Test Values")
plt.title("Confusion Matrix for CNN")
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn3=KNeighborsClassifier(n_neighbors=3)
knn3.fit(X,Y)
print(cross_val_score(knn3,X,Y,cv=10))
print(np.average(cross_val_score(knn3,X,Y,cv=10)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn4=KNeighborsClassifier(n_neighbors=5)
knn4.fit(X,Y)
print(cross_val_score(knn4,X,Y,cv=10))
print(np.average(cross_val_score(knn4,X,Y,cv=10)))
```
